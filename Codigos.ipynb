```python
import PySimpleGUI as sg
import pandas as pd

#define as perguntas pela planilha chamada "perguntas.xlsx"
resp = pd.read_excel('perguntas.xlsx')
p1 = resp.iloc[0,1]
p2 = resp.iloc[1,1]
p3 = resp.iloc[2,1]
p4 = resp.iloc[3,1]
p5 = resp.iloc[4,1]
p6 = resp.iloc[5,1]
p7 = resp.iloc[6,1]

r1 = resp.iloc[0,2]
r2 = resp.iloc[1,2]
resp.iloc[2,2] = 'Sigd'
r3 = resp.iloc[2,2]
r4 = resp.iloc[3,2]
r5 = resp.iloc[4,2]
r6 = resp.iloc[5,2]
r7 = resp.iloc[6,2]


#definição das funções de perguntas e intro

def intro():
    sg.theme('DarkAmber') 
    layout = [  [sg.Text('Deseja tentar desativar a bomba?')],
            [sg.Text('responda com cuidado as seguintes questões')],
            [sg.Button('Vamos nessa!'), sg.Button('sair')] ]
    window = sg.Window('intro', layout)
    
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Cancel':
            break
        else:
            window.close()
            return('passou')

def pergunta1():
        sg.theme('DarkAmber') 
        layout = [  [sg.Text(p1)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 1', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])
def pergunta2():
        sg.theme('Black') 
        layout = [  [sg.Text(p2)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 2', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])
            
def pergunta3():
        sg.theme('BlueMono') 
        layout = [  [sg.Text(p3)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')],#dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 3', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])            
            
def pergunta4():
        sg.theme('BluePurple') 
        layout = [  [sg.Text(p4)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 4', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])                       
            
def pergunta5():
        sg.theme('DarkBlue10') 
        layout = [  [sg.Text(p5)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 5', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])     
            
def pergunta6():
        sg.theme('LightBlue') 
        layout = [  [sg.Text(p6)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 6', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])   
            
def pergunta7():
        sg.theme('Kayak') 
        layout = [  [sg.Text(p7)],
                  [sg.Text('Dica: o arquivo de nome capital.mp3 lhes levará a localização de um item essencial para esta pergunta')], #dica
            [sg.Text('Digite aqui sua resposta (com maiúscula)'), sg.InputText()],
            [sg.Button('Próxima')] ]
        window = sg.Window('pergunta 7', layout)
        while True:
            event, values = window.read()
            if event == sg.WIN_CLOSED :
                break
            else:
                window.close()
                return(values[0])     
            
            
def parabens():
    sg.theme('BrightColors') 
    layout = [[sg.Text('PARABENS, BOMBA DESATIVADA COM SUCESSO')],
            [sg.Button('Fechar')] ]
    window = sg.Window('SUCESSO', layout)
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED :
            break            
    else:
        window.close()

#comandos        
if intro() == 'passou':
    while pergunta1() != r1:
        print('errou')
    while pergunta2() != 'r2':
        print('errou')
    
    while pergunta3()!=r3:
        print('errou')
    
    while pergunta4()!=r4:
        print('errou')
    while pergunta5()!=r5:
        print('errou')
    while pergunta6()!='r6':
        print('errou')
    while pergunta7()!=r7:
        print('errou')
    parabens()
    print('acertou')
```
